```python
from buildstock_query import UpgradesAnalyzer
```

```python
yaml_file = "project-file-example.yml"
buildstock_file = "buildstock.csv"
ua = UpgradesAnalyzer(yaml_file, buildstock_file)
```

```python
report_df = ua.get_report()
report_df
```

### Create upgraded buildstock

```python
upg = 1
df_bsl = ua.buildstock_df_original.set_index("Building")
df_upg = ua.get_upgraded_buildstock(upg).set_index("Building")

# sometimes it's easier to QC by comparing upgraded buildstock with baseline buildstock
print(f"Difference between upg {upg} buildstock and baseline : ")
df_diff = df_bsl.compare(df_upg, result_names=("baseline", "upgrade"))
df_diff
```
